```python
def create_strings_testing_char2(max_size: int, min_lcs_len: int):
    lcs = np.random.randint(0, 2, min_lcs_len)

    base_str = ''
    for c in lcs:
        base_str += str(c)

    rest = max_size - min_lcs_len
    prefix1_len = np.random.randint(0, rest - 1)
    prefix2_len = np.random.randint(0, rest - 1)

    prefix1 = ""
    suffix1 = ""

    prefix = np.random.randint(0, 2, prefix1_len)
    suffix = np.random.randint(0, 2, rest - prefix1_len - 1)

    for c in prefix:
        prefix1 += str(c)

    for c in suffix:
        suffix1 += str(c)

    prefix2 = ""
    suffix2 = ""
    prefix = np.random.randint(0, 2, prefix2_len)
    suffix = np.random.randint(0, 2, rest - prefix2_len - 1)

    for c in prefix:
        prefix2 += str(c)

    for c in suffix:
        suffix2 += str(c)

    s1 = prefix1+base_str+suffix1+'2'
    s2 = prefix2+base_str+suffix2+'3'
    
    return s1, s2


def map_strings(input_string: str):
    mapping = ['01','10','00','11']
    out = ""
    for i in input_string:
        out += mapping[int(i)]
    return out 

s1, s2 = create_strings_testing_char2(16, 2)

s1 = map_strings(s1)
s2 = map_strings(s2)

classical_lcs_substring_len(s1, s2)


def random_strings_testing_char2(size: int):
    base_str = ''
    seq = np.random.randint(0, 2, size-1)
    for c in seq:
        base_str += str(c)
    s1 = base_str+'2'

    base_str = ''
    seq = np.random.randint(0, 2, size-1)
    for c in seq:
        base_str += str(c)
        
    s2 = base_str+'3'
    return s1, s2

create_strings_testing_char2(16, 0)
```

```python
AMOUNT = 500

good = 480
bad = 20
good_strings = []
bad_strings = []


testing_dataset_short = []

while len(testing_dataset_short) == 100:
    x, y = create_strings_testing_char2(8, 0)
    x = map_strings(x)
    y = map_strings(y)
    if classical_lcs_substring_len(x, y) == 2:
        testing_dataset_short.append((x, y))

for x, y in testing_dataset_short:
    classical_result = classical_lcs_substring_len(x, y)
    quantum_result = quantum_LCS(x, y, len(x)//2, 2, verbose=False)
    
    if classical_result == quantum_result:
        good += 1
        good_strings.append((x,y))
        print(f"HIT:{good}, {x}, {y}")
        
    else:
        bad += 1
        bad_strings.append((x,y))
        print(f"MISS:{bad}, {x}, {y}")
    print(f"expected: {classical_result}, found {quantum_result}")
```
