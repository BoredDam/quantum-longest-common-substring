```python
AMOUNT = 36 # quanti test per lunghezza di LCS
count = 805 # numero da cui iniziare a indicizzare i test

# puoi definire la dimensione minima e massima su cui verranno effettuati i test
# se LCS_min_size = LCS_max_size, verranno effettuati solo i test su quella dimensione
LCS_min_size = 8
LCS_max_size = 8

for i in [13]:
    testing_dataset_short = []

    while len(testing_dataset_short) < AMOUNT:
        x, y = random_strings_testing_char2(16)
        x = map_strings(x)
        y = map_strings(y)
        if classical_lcs_substring_len(x, y) == i and (x, y) not in testing_dataset_short: # esclude ripetuti
            testing_dataset_short.append((x, y))


    for x, y in testing_dataset_short:
        classical_result = classical_lcs_substring_len(x, y)
        quantum_result = quantum_LCS(x, y, len(x)//2, 2, verbose=False)
        
        if classical_result == quantum_result:
            
            good_strings.append((x,y))
            print(f"HIT,{count},{x},{y},{classical_result},{quantum_result}")
            
        else:
            bad += 1
            bad_strings.append((x,y))
            print(f"MISS,{count},{x},{y},{classical_result},{quantum_result}")

        count += 1
```
