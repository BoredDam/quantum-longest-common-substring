# **Testing playground for the Quantum LCS algorithm - DNA 🧬**

The Longest Common Substring Algorithm is one of the most important algorithms in the field of bioinformatics, especially in DNA arrangement examination.

In this notebook we'll explore how the DNA matching problem can be traced back to the Longest Common Substring problem, coded in a proper format and solved using our hybrid Quantum Longest Common Substring algorithm.

## **Longest Common Substring in DNA studying 👩🏻‍🔬👨🏻‍💻**

In bioinformatics, string matching tools are foundational tools, since DNA is essentially a massive "text" composed of four characters: A, C, G, and T.
Both the Longest Common Substring and the Longest Common Subsequence solutions, can be used to compare similarity between two DN sequences $X$ and $Y$.

### **Creating our data**

First, we are going to create some fake DNA sequences. You should pick a length which is near but less of a power of 2. We are going to see later why.

```python
SEQ_LEN_1 = 15
SEQ_LEN_2 = 15

def dna_random_strings(seq_len: int):
    """
    returns a DNA sequence of size `seq_len`
    """
    rand = np.random.randint(0, 4, size=seq_len)
    mapping = ['A','G','C','T' ]
    out = ""

    for r in rand:
        out += mapping[r]
    return out

seq1 = dna_random_strings(SEQ_LEN_1)
seq2 = dna_random_strings(SEQ_LEN_2)

print(f"sequence 1: {seq1}, sequence 2: {seq2}")
```

### **Adapting our strings**

Then, we'll adapt our strings to be used as input for our algorithm. Since our algorithm can only work on strings of size power of 2, and they also need termination characters (different between strings), we have to specify the output size.

```python
MAX_LEN = 16

def fix_shape_dna_lcs_problem(dna1: str, dna2: str, out_len: int):
    """
    retuns the given strings adapted for the QuantumLCS algorithm
    """
    input_len1 = len(dna1)
    input_len2 = len(dna2)
    dna1 = dna1 + '$'*(out_len - input_len1)
    dna2 = dna2 + '%'*(out_len - input_len2)
    return dna1, dna2

X, Y = fix_shape_dna_lcs_problem(seq1, seq2, MAX_LEN)

print(f"sequence 1: {X}, sequence 2: {Y}")
```

### **Mapping to binary codes**

The quantum circuit involved in our algorithm can't obviously work on characters like $A$, $C$, $G$ and $T$, but only on binary values that can be encoded in qubits, such as $000, 001, 010, 011, \dots$. In general, we need to encode our characters in strings $\in \{0,1\}^3$

```python
def dna_to_lcs_problem(dna_seq1: str, dna_seq2: str):
    """
    returns the DNA sequence mapped to a binary alphabet
    """
    mapping = {
    'A': '001', 'G': '010',
    'C': '011', 'T': '100',
    '$': '000', '%': '111'
    }

    out1 = ""
    out2 = ""
    for c in dna_seq1:
        out1 += mapping.get(c)
    for c in dna_seq2:
        out2 += mapping.get(c)
    return out1, out2

X_bin, Y_bin = dna_to_lcs_problem(X, Y)

print(f"sequence 1: {X_bin}, sequence 2: {Y_bin}")
```

## **Using our algorithm ⚛️🖥️**

We are now ready to use our hybrid Quantum Algorithm on the DNA sequences we just adapted. 
We'll compare our result to the one of a classical LCS algorithm.

```python
expected = classical_lcs_substring_len(X_bin, Y_bin, 3)
found = quantum_LCS(X_bin, Y_bin, MAX_LEN, 3, False)

if found == expected:
    print(f"{seq1} and {seq2} have a longest common substring of length {found}!")
else:
    print(f"unlucky! \nexpected result: {expected}, \nour result: {found}.")
```
